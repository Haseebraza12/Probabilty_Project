```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = "Youth Migration Trends - Form Responses 1.csv"
df = pd.read_csv(file_path)

# Count the occurrences of each gender
gender_counts = df['Your gender:'].value_counts()

# Calculate the total sample size and percentages
total_responses = gender_counts.sum()
male_percentage = (gender_counts.get('Male', 0) / total_responses) * 100
female_percentage = (gender_counts.get('Female', 0) / total_responses) * 100

# Plot the bar chart
plt.figure(figsize=(8, 6))
plt.bar(gender_counts.index, gender_counts.values, color=['skyblue', 'salmon', 'lightgreen'])
plt.xlabel("Gender")
plt.ylabel("Number of Respondents")
plt.title("Gender Distribution of Respondents")

# Add text with total sample size and male/female participation percentages
plt.text(2, total_responses * 0.9, f'Total Sample Size: {total_responses}', ha='center', fontsize=12, color='black')
plt.text(2, total_responses * 0.85, f'Male Participation: {male_percentage:.1f}%', ha='center', fontsize=12, color='blue')
plt.text(2, total_responses * 0.8, f'Female Participation: {female_percentage:.1f}%', ha='center', fontsize=12, color='red')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = "Youth Migration Trends - Form Responses 1.csv"
df = pd.read_csv(file_path)

# Convert the 'Your age:' column to numeric, handling non-numeric values as NaN
df['Your age:'] = pd.to_numeric(df['Your age:'], errors='coerce')

# Drop NaN values from the age column
age_data = df['Your age:'].dropna()

# Create bins for age groups
bins = [15, 20, 25, 30, 35, 40]  # Adjust bins as needed
age_data_binned = pd.cut(age_data, bins=bins)

# Count the occurrences in each age bin
age_counts = age_data_binned.value_counts().sort_index()

# Plot the age distribution as a bar chart
plt.figure(figsize=(12, 6))
age_counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.xlabel("Age Groups")
plt.ylabel("Number of Respondents")
plt.title("Age Distribution of Respondents")
plt.xticks(rotation=45)

plt.show()
```

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

# Load the CSV file
file_path = "Youth Migration Trends - Form Responses 1.csv"
df = pd.read_csv(file_path)

# Helper function for creating bar charts
def plot_bar_chart(data, title):
    fig = px.bar(data, x=data.index, y=data.values, labels={'x': 'Response', 'y': 'Count'}, title=title)
    fig.show()

# Helper function for creating pie charts
def plot_pie_chart(data, title):
    fig = px.pie(data, names=data.index, values=data.values, title=title)
    fig.show()

# Question 3: Highest level of education completed by respondents
education_counts = df['What is the highest level of education you have completed? '].value_counts()
plot_bar_chart(education_counts, "Highest Level of Education Completed by Respondents")

# Question 4: Have respondents considered studying abroad?
abroad_interest_counts = df['Have you ever considered studying abroad? '].value_counts()
plot_pie_chart(abroad_interest_counts, "Interest in Studying Abroad")

# Question 5: Level of education considered for studying abroad (for those who answered yes)
education_abroad_counts = df['  If yes, what level of education are you considering studying abroad?  '].dropna().value_counts()
plot_bar_chart(education_abroad_counts, "Level of Education Considered for Studying Abroad")

# Question 6: Motivations for studying abroad
motivations_abroad_counts = df['What has motivated you to consider studying abroad? (Select all that apply)  '].str.get_dummies(', ').sum()
plot_bar_chart(motivations_abroad_counts, "Motivations for Considering Studying Abroad")

# Question 7: Impact of educational opportunities in Pakistan on career goals
education_impact_counts = df['Which of the following best describes how educational opportunities in Pakistan have impacted your career goals? '].value_counts()
plot_pie_chart(education_impact_counts, "Impact of Educational Opportunities in Pakistan on Career Goals")

# Question 8: Preference for studying in Pakistan or abroad
study_preference_counts = df['If given the chance, would you choose to study at a university in Pakistan or abroad, assuming both offer equal value in terms of education, opportunities, and scholarships?  '].value_counts()
plot_pie_chart(study_preference_counts, "Preference for Studying in Pakistan or Abroad")

# Question 9: Challenges of life in Pakistan
challenges_in_pakistan_counts = df['Which aspects of life in Pakistan do you find most challenging? (Select all that apply)  '].str.get_dummies(', ').sum()
plot_bar_chart(challenges_in_pakistan_counts, "Challenges of Life in Pakistan")

# Question 10: Appeals of living abroad
appeals_abroad_counts = df['What do you find most appealing about living abroad? (Select all that apply)  '].str.get_dummies(', ').sum()
plot_bar_chart(appeals_abroad_counts, "Appeals of Living Abroad")

# Question 11: Perception of growth opportunities in Pakistan
growth_opportunities_counts = df['Do you think there are enough opportunities for personal and professional growth in Pakistan? '].value_counts()
plot_pie_chart(growth_opportunities_counts, "Perception of Growth Opportunities in Pakistan")

# Question 12: Plan to seek employment opportunities abroad
employment_abroad_counts = df['Do you plan to seek employment opportunities abroad? '].value_counts()
plot_pie_chart(employment_abroad_counts, "Plan to Seek Employment Opportunities Abroad")

# Question 13: Competitive salary in Pakistan and migration consideration
salary_migration_counts = df['If you were offered a competitive salary in Pakistan, would you consider migrating abroad for work?  '].value_counts()
plot_pie_chart(salary_migration_counts, "Migration Consideration with Competitive Salary in Pakistan")

# Question 14: Effect of political and economic situation in Pakistan
politics_effect_counts = df['How does the political and economic situation in Pakistan affect your thoughts on leaving the country? '].value_counts()
plot_bar_chart(politics_effect_counts, "Effect of Political and Economic Situation on Migration Thoughts")

# Question 15: Influence of achievements of people abroad
inspiration_abroad_counts = df['Do you know anyone abroad whose achievements or lifestyle has inspired you to move to another country?  '].value_counts()
plot_pie_chart(inspiration_abroad_counts, "Influence of Achievements of People Abroad")

# Question 16: Countries considered for moving
countries_considered_counts = df['Which countries have you thought about moving to? (Select all that apply)  '].str.get_dummies(', ').sum()
plot_bar_chart(countries_considered_counts, "Countries Considered for Moving Abroad")

# Question 17: Challenges anticipated when moving abroad
challenges_abroad_counts = df['What do you think is the biggest challenge of moving abroad?  (Select all that apply)     '].str.get_dummies(', ').sum()
plot_bar_chart(challenges_abroad_counts, "Anticipated Challenges of Moving Abroad")

# Question 18: Impact of moving abroad on marriage decision
marriage_impact_counts = df['If you were to move abroad, how would it affect your decision on where to marry?  '].value_counts()
plot_pie_chart(marriage_impact_counts, "Impact of Moving Abroad on Marriage Decision")
```

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import plotly.express as px

# Assuming the DataFrame is loaded as df
# Question: Highest Level of Education Completed
education_counts = df['What is the highest level of education you have completed? '].value_counts()

# Plotting Bar Chart
fig = px.bar(education_counts, x=education_counts.index, y=education_counts.values,
             labels={'x': 'Education Level', 'y': 'Count'}, title="Education Level of Youth in Pakistan")
fig.show()

# Calculating Confidence Interval for Education Level (Assuming it's a numerical scale)
education_numeric = df['What is the highest level of education you have completed? '].map({
    'Primary': 1, 'Secondary': 2, 'Tertiary': 3, 'Undergraduate': 4, 'Postgraduate': 5
})

# Confidence Interval for the average education level
mean = education_numeric.mean()
std_dev = education_numeric.std()
n = len(education_numeric)
confidence_interval = stats.t.interval(0.95, n-1, loc=mean, scale=std_dev/np.sqrt(n))

# Output the Results
print(f"Confidence Interval for Education Level (95% CI): {confidence_interval}")
```

```python
# Question: Have you considered studying abroad?
abroad_interest_counts = df['Have you ever considered studying abroad? '].value_counts()

# Plotting Pie Chart for Studying Abroad Consideration
fig = px.pie(abroad_interest_counts, values=abroad_interest_counts.values, names=abroad_interest_counts.index,
             title="Youth Considering Studying Abroad")
fig.show()

# Calculating Probability of Youth Considering Studying Abroad
probability_abroad = abroad_interest_counts['Yes'] / abroad_interest_counts.sum()
print(f"Probability of Youth Considering Studying Abroad: {probability_abroad}")
```

```python
# Question: What has motivated you to consider studying abroad?
motivation_counts = df['What has motivated you to consider studying abroad? (Select all that apply) '].value_counts()

# Plotting Bar Chart for Motivation to Study Abroad
fig = px.bar(motivation_counts, x=motivation_counts.index, y=motivation_counts.values,
             labels={'x': 'Motivation', 'y': 'Count'}, title="Motivation for Studying Abroad")
fig.show()

# Calculating Probability Distribution for Motivations
motivation_prob = motivation_counts / motivation_counts.sum()
print(f"Motivation Probability Distribution: \n{motivation_prob}")
```
